```python
# imports

import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# Ruta del archivo CSV
file_path = '/content/drive/MyDrive/Colab Notebooks/Copia de F1DriversDataset.csv' #Ruta Joaquín Cerdan

# Leo el archivo CSV en un DataFrame de Pandas
f1_df = pd.read_csv(file_path)

# Conexión a la base de datos en memoria
conn = sqlite3.connect(':memory:')

# Escribir el DataFrame a una tabla SQLite
f1_df.to_sql('f1_dataset', conn, index=False, if_exists='replace')
```

## Lectura de datos (SQL)



```python
# Mostrar los primeros 5 registros
query1 = "SELECT * FROM f1_dataset LIMIT 5"
df_query1 = pd.read_sql(query1, conn)
df_query1
```

```python
#Se muestran las columnas mas relevantes de los ulitmos 10 registros (order by Years_Active)
query2 = "SELECT Driver, Nationality, Championships, Race_Wins, Decade, Champion FROM f1_dataset ORDER BY Years_Active DESC LIMIT 10"
df_query2 = pd.read_sql(query2, conn)
df_query2
```

```python
# Mostrar los 10 pilotos con mas carrearas ganadas
query3 = "SELECT * FROM f1_dataset ORDER BY Race_Wins DESC LIMIT 10"
df_query3 = pd.read_sql(query3, conn)
df_query3
```

```python
# Mostrar 10 nacionalidades que mas aparecen en F1
query4 = "SELECT Nationality, COUNT(Nationality) as Total_Nationality FROM f1_dataset GROUP BY Nationality ORDER BY Total_Nationality DESC LIMIT 10"
df_query4 = pd.read_sql(query4, conn)
df_query4
```

```python
# Mostrar 5 piolotos con mas campeonatos
query5 = "SELECT Driver, SUM(Championships) as Total_Championships FROM f1_dataset GROUP BY Driver ORDER BY Total_Championships DESC LIMIT 5"
df_query5 = pd.read_sql(query5, conn)
df_query5
```

```python
# Mostrar el registro de Michael Schumacher
query6 = "SELECT * FROM f1_dataset WHERE driver = 'Michael Schumacher'"
df_query6 = pd.read_sql(query6, conn)
df_query6
```

## Análisis univariado y bivariado + visualización de datos (Python - Matplotlib/Seaborn)

Histograma de los años activos de cada piloto

```python
plt.figure(figsize=(7, 5))
ax = sns.histplot(f1_df['Years_Active'], kde=True, bins=30)
ax.set_facecolor('#E0E0E0') #Se cambia el color del fondo para una mejor visualizacion
plt.title('Distribución años activos de pilotos')
plt.xlabel('Años activos')
plt.ylabel('Frecuencia')
plt.show()
```

```python
def mostrar_histograma(columna): #Se define una funcion para poder mostrar varios histogramas

  plt.figure(figsize=(10, 8))
  ax = sns.histplot(f1_df[columna], kde=True, bins=30)
  ax.set_facecolor('#E0E0E0') #Se cambia el color del fondo para una mejor visualizacion
  plt.xticks(rotation=90) #Rotar las etiquetas del eje x para que se vean bien
  plt.title(f'Distribución de {columna} de los Pilotos')
  plt.xlabel(columna)
  plt.ylabel('Frecuencia')
  plt.show()
```

```python
mostrar_histograma("Race_Wins")
```

La GRAN mayoria de los pilotos NUNCA ganan una carreara de F1 en toda su carrera

```python
mostrar_histograma("Race_Starts")
```

```python
mostrar_histograma("Decade")
```

Se puede observar como en 1960 hubo la mayor cantidad de pilotos

```python
# Scatter plot de la relación entre los años activos y las carrearas corridas
plt.figure(figsize=(8, 8))
ax = sns.scatterplot(x='Years_Active', y='Race_Starts', hue='Active', data=f1_df)
ax.set_facecolor('#E0E0E9')
plt.title('Relación entre los años activos y las carrearas corridas (Activos en rojo)')
plt.xlabel('Años activo')
plt.ylabel('Carreras corridas')
plt.show()
```

```python
import numpy as np
numerical_columns = f1_df.select_dtypes(include=[np.number])

corr_matrix = numerical_columns.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()
```

## Conclusiones & insights


**El mejor piloto de la historia**

Basándonos en carreras ganadas, campeonatos obtenidos y puntos promedio por carrera, Lewis Hamilton destaca como el mejor piloto de Fórmula 1 de todos los tiempos. Es seguido por Michael Schumacher, quien también ha dejado un legado extraordinario en la categoría. El tercer lugar del podio puede ser motivo de debate entre otros grandes pilotos como Juan Manuel Fangio, Alain Prost, Ayrton Senna y Sebastian Vettel, quienes marcaron épocas y contribuyeron significativamente al prestigio de la Fórmula 1.

**Relaciones clave entre las métricas de desempeño**

Existe una relación muy directa entre el número de pole positions, victorias en carrera y podios alcanzados. Esto es algo lógico considerando que obtener una buena posición en la parrilla de salida (pole position) aumenta considerablemente las probabilidades de ganar carreras o alcanzar un podio. Este patrón refuerza la importancia del rendimiento constante y la excelencia en la clasificación para tener éxito en la Fórmula 1.

**La exclusividad y dificultad de la Fórmula 1**

Los histogramas del análisis reflejan cómo la Fórmula 1 es el evento automovilístico más importante y exclusivo del mundo. Muy pocos pilotos logran siquiera sentarse en un monoplaza de Fórmula 1 y, de aquellos que lo hacen, solo un porcentaje mínimo consigue destacarse con victorias, podios o títulos mundiales.
Además, mantener una carrera longeva en la Fórmula 1 es extremadamente difícil y requiere un nivel excepcional de habilidad y consistencia. La mayoría de los registros en el dataset corresponden a pilotos que participaron en algunas carreras pero no lograron dejar una huella significativa. Este hecho resalta lo extraordinario que es triunfar y mantenerse relevante en la élite del automovilismo durante varios años.

***Conclusión***

El analisis de este dataset confirman que la Fórmula 1 no solo es un espectáculo de velocidad, sino también un escenario que exige un talento extraordinario, consistencia y resiliencia. Su exclusividad y nivel competitivo la posicionan como el pináculo del automovilismo a nivel mundial.
